Credit card fraud is a major issue in financial systems, where unauthorized transactions lead to financial loss.

This project aims to analyze the dataset using Exploratory Data Analysis (EDA) techniques to understand patterns, trends, and relationships between variables that can help in detecting fraudulent activities.

**4.Data Visualization, Storytelling & Experimenting With Charts: Relationship Between Variables**

**Chart-1**

```python
#Fraud vs Normal Transactions chart to understand class imbalance.
df['Class'].value_counts().plot(kind='bar', title='Fraud vs Normal Transactions')
```

What are the insights that you found from the chart?



*   Normal transactions are much more than fraud transactions.
*   Fraud cases are very rare.








**Chart-2**

```python
#Average Amount by Class chart tells us whether fraudulent transactions usually involve higher money.
df.groupby('Class')['Amount'].mean().plot(kind='bar', title='Average Amount by Class')
```

What are the insights that you found from the chart?





*   Fraud transactions may have different average amount compared to normal ones.
*  This helps us see if fraud involves more or less money.



**Chart-3**

```python
#Transaction Time Distribution helps to identify peak hours
df['Time'].plot(kind='hist', bins=30, title='Transaction Time Distribution')
```

What are the insights that you found from the chart?


*  Transactions are spread over time.
*   Some time periods have more transactions than others.



**Chart-4**

```python
#Amount vs Class chart helps visualize at what amounts fraud happens
df.plot(kind='scatter', x='Amount', y='Class', title='Amount vs Class')
```

What are the insights that you found from the chart?



*   Fraud can happen at any transaction amount.
*   It is not limited to only high or low values.



**Chart-5**

```python
#Transaction Amount Distribution chart shows how transaction amounts are spread.
df['Amount'].plot(kind='hist', bins=50, title='Transaction Amount Distribution')
```

What are the insights that you found from the chart?



*   Most transactions are small amounts.
*   Very large transactions are rare.



**Chart**-**6**

```python
#Fraud Transaction Amount Distribution shows only fraud amounts.
df[df['Class'] == 1]['Amount'].plot(kind='hist', bins=30, title='Fraud Transaction Amount Distribution')
```

What are the insights that you found from the chart?



*   Fraud transactions also occur mostly in certain amount ranges.
*   Helps identify common fraud amount patterns.



**Chart-7**

```python
#This chart is used to find outliers
df.boxplot(column='Amount', by='Class')
```

What are the insights that you found from the chart?



*   Shows difference in spread of amounts between fraud and normal.
*   Helps detect outliers (very unusual transactions).



**Chart-8**

```python
#Transaction Time Trend shows how transactions happen over time.
df['Time'].head(1000).plot(kind='line', title='Transaction Time Trend (First 1000 Rows)')
```

What are the insights that you found from the chart?



*   Transactions happen continuously over time.
*   Some periods may have more activity (busy times).



**Chart-9**

```python
#This chart is used to show relationship between tables
df[['Time', 'Amount', 'Class']].corr().style.background_gradient()
```

What are the insights that you found from the chart?



*   Relationship between variables is weak.
*   Fraud is not strongly dependent on only time or amount.


**Chart-10**

```python
#This chart is used to show highest fraud amounts
df[df['Class'] == 1]['Amount'].sort_values(ascending=False).head(10).plot(kind='bar', title='Top 10 Highest Fraud Amounts')
```

What are the insights that you found from the chart?



*   Shows the largest fraud transactions.
*   Helps understand maximum loss due to fraud.

This project focused on analyzing credit card transaction data to understand fraud patterns using Exploratory Data Analysis (EDA).
